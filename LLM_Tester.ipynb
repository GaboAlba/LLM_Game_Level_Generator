```python
import os
import openai
from google import genai

# Get the different API keys from the repo's secrets
OpenAi_API_Key = os.environ["OPEN_AI_API_KEY"]
Gemini_API_Key = os.environ["GEMINI_API_KEY"]

class LLMClient:
  def _init_(self, api_key : str, model : str, allowedModels : list):
    self.api_key = api_key
    self.model = model
    self.allowedModels = allowedModels
    if self.model not in self.allowedModels:
      raise ValueError(f"Model {self.model} is not allowed. Allowed models are: {self.allowedModels}")

class OpenAIClient(LLMClient):  
  def _init_(self, api_key : str, model : str):
    self.allowedModels = openai.models.list()
    super()._init_(api_key, model)
    self.client = openai.OpenAI(api_key= self.api_key, model = self.model)

class GeminiClient(LLMClient):
  def _init_(self, api_key : str, model : str):
    self.allowedModels = self.client.models.list()
    super()._init_(api_key, model)
    self.client = genai.Client(api_key=self.api_key, model=self.model)

```
